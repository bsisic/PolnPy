## Analysis on Alnus

Plot the pollen period as average over the years

```python
aln = pol_wu_df[['DayOfYear', 'Alnus']].groupby('DayOfYear').mean().reset_index()
```

```python
aln.head()
```

```python
ax = aln.plot(x='DayOfYear',y='Alnus',colormap='Spectral')
ax.set_title('Average concentration over 22 years',fontsize=14, fontweight='bold')
ax.set_ylabel("Grains/m3")
```

```python
alns = pol_wu_df[['Year','DayOfYear', 'Alnus']]
alns = alns[(alns['DayOfYear']>0) & (alns['DayOfYear']<125)]
```

```python
fg = sns.FacetGrid(alns, col="Year", col_wrap=5, size=2.5)
fg = fg.map(plt.plot, "DayOfYear", "Alnus")
plt.subplots_adjust(top=0.9)
fg.fig.suptitle('Alnus concentration over 22 years (day #0-125)',fontsize=14, fontweight='bold')
```

```python
sns.set_palette("GnBu_d")
sns.set_style('whitegrid')
jp1 = sns.jointplot(x='DewPointAvg',y='Alnus',
              data=pol_wu_df[(pol_wu_df['DayOfYear']>0) & (pol_wu_df['DayOfYear']<125)],
              kind='hex',bins=10)
jp1.fig.suptitle('Correlation Alnus and Dew Point (day #0-150)', fontsize=14, fontweight='bold')
plt.tight_layout()
```

```python
jp2 = sns.jointplot(x='WindAvg',y='Alnus',
                  data=pol_wu_df[(pol_wu_df['DayOfYear']>0) & (pol_wu_df['DayOfYear']<125)],
                  kind='hex',bins=5)
jp2.fig.suptitle('Correlation Alnus and Wind between day #0 and day #150', fontsize=14, fontweight='bold')
plt.tight_layout()
```
